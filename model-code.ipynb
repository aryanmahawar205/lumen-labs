Task 1 - Basic Image Statistics and Color Space Conversion


```python
#1. Read the Image: Load an image using OpenCV.
import cv2

image_path = '/content/Ferrari-Logo.png'
image = cv2.imread(image_path)
print(image)
```

```python
#2. Compute Basic Statistics: Calculate the mean, standard deviation, and histogram of each color channel.
import numpy as np
import matplotlib.pyplot as plt
blue_channel, green_channel, red_channel = cv2.split(image)

# Calculate mean and standard deviation for each channel
blue_mean, blue_stddev = cv2.meanStdDev(blue_channel)
green_mean, green_stddev = cv2.meanStdDev(green_channel)
red_mean, red_stddev = cv2.meanStdDev(red_channel)

print(f'Blue Channel - Mean: {blue_mean[0][0]}, StdDev: {blue_stddev[0][0]}')
print(f'Green Channel - Mean: {green_mean[0][0]}, StdDev: {green_stddev[0][0]}')
print(f'Red Channel - Mean: {red_mean[0][0]}, StdDev: {red_stddev[0][0]}')

# Calculate histograms for each channel
blue_hist = cv2.calcHist([blue_channel], [0], None, [256], [0, 256])
green_hist = cv2.calcHist([green_channel], [0], None, [256], [0, 256])
red_hist = cv2.calcHist([red_channel], [0], None, [256], [0, 256])

# Plot histograms for each channel
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.title('Blue Channel Histogram')
plt.plot(blue_hist, color='blue')
plt.xlim([0, 256])

plt.subplot(1, 3, 2)
plt.title('Green Channel Histogram')
plt.plot(green_hist, color='green')
plt.xlim([0, 256])

plt.subplot(1, 3, 3)
plt.title('Red Channel Histogram')
plt.plot(red_hist, color='red')
plt.xlim([0, 256])

plt.show()
```

```python
#3. Convert Color Spaces: Convert the image to HSV and Lab color spaces and display the results.
# Convert the image to HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Convert the image to Lab color space
lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2Lab)

# Display the original and converted images
plt.figure(figsize=(18, 6))

# Display the original image
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Display the HSV image
plt.subplot(1, 3, 2)
plt.title('HSV Image')
plt.imshow(cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB))
plt.axis('off')

# Display the Lab image
plt.subplot(1, 3, 3)
plt.title('Lab Image')
plt.imshow(cv2.cvtColor(lab_image, cv2.COLOR_Lab2RGB))
plt.axis('off')

plt.show()
```

Task 2: Simple Image Segmentation Using Thresholding

```python
#1. Read the Image: Load a grayscale image.
# Load a grayscale image using OpenCV
image_path = '/content/Ferrari-Logo.png'
grayscale_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display the grayscale image
plt.figure(figsize=(6, 6))
plt.title('Grayscale Image')
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#2. Apply Thresholding: Use a fixed threshold value to segment the image.
# Apply fixed thresholding
threshold_value = 128
_, thresholded_image = cv2.threshold(grayscale_image, threshold_value, 255, cv2.THRESH_BINARY)
```

```python
#3. Display Results: Show the original and segmented images.
# Display the original and thresholded images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Grayscale Image')
plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Thresholded Image')
plt.imshow(thresholded_image, cmap='gray')
plt.axis('off')

plt.show()
```

Task 3: Color-Based Segmentation

```python
#1. Read the Image: Load an image with objects of different colors.
# Load an image using OpenCV
image_path = '/content/Ferrari-Logo.png'
img_a = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Convert BGR image to RGB for displaying with matplotlib
img_b = cv2.cvtColor(img_a, cv2.COLOR_BGR2RGB)

# Display the color image
plt.figure(figsize=(8, 8))
plt.title('Color Image')
plt.imshow(img_b)
plt.axis('off')
plt.show()
```

```python
#2. Convert to HSV: Convert the image to HSV color space.
# Convert BGR image to HSV
img_c = cv2.cvtColor(img_a, cv2.COLOR_BGR2HSV)

# Display the original color image and the HSV image
plt.figure(figsize=(12, 6))

# Convert BGR to RGB for proper display with matplotlib
img_b = cv2.cvtColor(img_a, cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.title('Original Color Image')
plt.imshow(img_b)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('HSV Image')
plt.imshow(img_c)
plt.axis('off')

plt.show()
```

```python
#3. Apply Color Thresholding: Use color thresholds to segment objects of a specific color.
# Convert the image from BGR to RGB (OpenCV loads images in BGR format)
img_b = cv2.cvtColor(img_a, cv2.COLOR_BGR2RGB)

# Convert the image to HSV color space
img_c = cv2.cvtColor(img_a, cv2.COLOR_BGR2HSV)

# Define the color range for segmentation
lower_red = np.array([45, 45, 45])
upper_red = np.array([100, 255, 255])


# Create a mask using the color range
mask_e = cv2.inRange(img_c, lower_red, upper_red)

# Apply the mask to the original image
img_f = cv2.bitwise_and(img_b, img_b, mask=mask_e)
```

```python
#4. Display Results: Show the original and segmented images.
# Display the original RGB image
plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.title('Original Image')
plt.imshow(img_b)
plt.axis('off')

# Display the mask
plt.subplot(2, 3, 4)
plt.title('Mask')
plt.imshow(mask_e, cmap='gray')
plt.axis('off')

# Display the segmented image
plt.subplot(2, 3, 5)
plt.title('Segmented Image')
plt.imshow(img_f)
plt.axis('off')

plt.show()
```

Task 4: Feature Extraction from Segmented Objects

```python
import cv2
import matplotlib.pyplot as plt

# Load an image from file
image_path = '/content/Ferrari-Logo.png'
img_1 = cv2.imread(image_path)

# Convert the image to grayscale
gray_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to the image
blurred_1 = cv2.GaussianBlur(gray_1, (5, 5), 0)

# Perform thresholding to create a binary image
ret_1, binary_1 = cv2.threshold(blurred_1, 127, 255, cv2.THRESH_BINARY)

# Find contours of the segmented objects
contours_1, hierarchy_1 = cv2.findContours(binary_1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
img_with_contours_1 = cv2.drawContours(img_1.copy(), contours_1, -1, (0, 255, 0), 2)

# Display the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB))

plt.subplot(2, 2, 2)
plt.title('Grayscale Image')
plt.imshow(gray_1, cmap='gray')

plt.subplot(2, 2, 3)
plt.title('Binary Image')
plt.imshow(binary_1, cmap='gray')

plt.subplot(2, 2, 4)
plt.title('Segmented Image with Contours')
plt.imshow(cv2.cvtColor(img_with_contours_1, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
#2. Extract Features: Calculate shape, color, and texture features for each segmented object (e.g.,area, perimeter, mean color, texture metrics).
from skimage.feature import graycomatrix, graycoprops

# Load an image from file
image_path = '/content/Ferrari-Logo.png'
img_201 = cv2.imread(image_path)

# Convert the image to grayscale
gray_202 = cv2.cvtColor(img_201, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to the image
blurred_203 = cv2.GaussianBlur(gray_202, (5, 5), 0)

# Perform thresholding to create a binary image
ret_204, binary_205 = cv2.threshold(blurred_203, 127, 255, cv2.THRESH_BINARY)

# Find contours of the segmented objects
contours_206, hierarchy_207 = cv2.findContours(binary_205, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Function to calculate texture features using GLCM
def calculate_texture_features(gray_image, mask):
    glcm = graycomatrix(gray_image, distances=[5], angles=[0], levels=256, symmetric=True, normed=True)
    contrast = graycoprops(glcm, 'contrast')[0, 0]
    dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]
    correlation = graycoprops(glcm, 'correlation')[0, 0]
    return contrast, dissimilarity, homogeneity, energy, correlation

# Iterate over each contour and extract features
features_208 = []
for contour_209 in contours_206:
    # Calculate area and perimeter
    area_210 = cv2.contourArea(contour_209)
    perimeter_211 = cv2.arcLength(contour_209, True)

    # Create a mask for the current contour
    mask_212 = np.zeros_like(gray_202)
    cv2.drawContours(mask_212, [contour_209], -1, 255, thickness=cv2.FILLED)

    # Calculate mean color within the contour
    mean_color_213 = cv2.mean(img_201, mask=mask_212)

    # Calculate texture features
    contrast_214, dissimilarity_215, homogeneity_216, energy_217, correlation_218 = calculate_texture_features(gray_202, mask_212)

    # Store the features
    features_208.append({
        'area': area_210,
        'perimeter': perimeter_211,
        'mean_color': mean_color_213,
        'contrast': contrast_214,
        'dissimilarity': dissimilarity_215,
        'homogeneity': homogeneity_216,
        'energy': energy_217,
        'correlation': correlation_218
    })

```

```python
#3. Analyze Features: Display and interpret the extracted features
# Display the results
for i, feature_219 in enumerate(features_208):
    print(f"Object {i+1}:")
    print(f"  Area: {feature_219['area']}")
    print(f"  Perimeter: {feature_219['perimeter']}")
    print(f"  Mean Color (BGR): {feature_219['mean_color']}")
    print(f"  Contrast: {feature_219['contrast']}")
    print(f"  Dissimilarity: {feature_219['dissimilarity']}")
    print(f"  Homogeneity: {feature_219['homogeneity']}")
    print(f"  Energy: {feature_219['energy']}")
    print(f"  Correlation: {feature_219['correlation']}")
    print()
```

Task 5: Multi-Object Segmentation and Counting

```python
#1. Read the Image: Load an image containing multiple objects.

image_path = '/content/Ferrari-Logo.png'
e = cv2.imread(image_path)

# Convert the image from BGR to RGB for displaying correctly with matplotlib
e_rgb = cv2.cvtColor(e, cv2.COLOR_BGR2RGB)

# Display the image
plt.figure(figsize=(8, 6))
plt.title('Loaded Image')
plt.imshow(e_rgb)
plt.axis('off')
plt.show()
```

```python
#2. Preprocess the Image: Apply preprocessing techniques such as filtering and color space conversion.

# Convert the image to grayscale
g = cv2.cvtColor(e, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to the grayscale image
z = cv2.GaussianBlur(g, (5, 5), 0)

# Convert the original image to the HSV color space
l = cv2.cvtColor(e, cv2.COLOR_BGR2HSV)

# Display the results
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(e, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('Grayscale Image')
plt.imshow(g, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title('Gaussian Blurred Image')
plt.imshow(z, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title('HSV Color Space Image')
plt.imshow(cv2.cvtColor(l, cv2.COLOR_HSV2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#3. Segment Objects: Use techniques like thresholding, contour detection, or clustering to segment the objects.
# Convert the image to grayscale
g = cv2.cvtColor(e, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to the grayscale image
z = cv2.GaussianBlur(g, (5, 5), 0)

# Perform thresholding to create a binary image
_, l = cv2.threshold(z, 127, 255, cv2.THRESH_BINARY)

# Find contours of the segmented objects
contours, _ = cv2.findContours(l, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
e_with_contours = cv2.drawContours(e.copy(), contours, -1, (0, 255, 0), 2)

# Display the results
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(e, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title('Gaussian Blurred Image')
plt.imshow(z, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title('Segmented Image with Contours')
plt.imshow(cv2.cvtColor(e_with_contours, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#4. Count Objects: Identify and count the number of segmented objects.

# Count the number of segmented objects
number_of_objects = len(contours)
print(f"Number of segmented objects: {number_of_objects}")
```

```python
#5. Display Results: Show the original image with the segmented objects highlighted and the count of objects.
# Count the number of segmented objects
number_of_objects = len(contours)

plt.subplot(2, 2, 4)
plt.title(f'Segmented Image with Contours (Count: {number_of_objects})')
plt.imshow(cv2.cvtColor(e_with_contours, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()

print(f"Number of segmented objects: {number_of_objects}")
```
